```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv("diabetes.csv")
# df = df.fillna(df.mean())
print("NaN value exists?:", df.isnull().values.any())
df.head()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
len(df.columns)
```

```python
for col in df.columns:
    sns.lmplot(col, df.columns[-1], data=df, ci = None, fit_reg=False)
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
def get_imputed_data(option, train, test):
    print("-----------------------------------------")
    print("Imputer:", option)
    imputer = Imputer(missing_values="NaN", strategy=option, axis=0)
    train_x_imputed = imputer.fit_transform(train)
    test_x_imputed = imputer.fit_transform(test)
    return train_x_imputed, test_x_imputed
```

```python
def runClassifier(clf, tr_x, tr_y, ts_x, ts_y):
    clf.fit(tr_x, tr_y)
    hyp = clf.predict(ts_x)
    score = accuracy_score(ts_y, hyp)
    print("Accuracy: %0.4f" % (score))
    return score
```

```python
from sklearn.preprocessing import Imputer

imputer_opt = ["most_frequent", "median", "mean"]
classifier_opt = {
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier()
}
best_score = 0
best_model = None
best_imputer = None

for name, clf in classifier_opt.items():
        print("Model:", name)
        for imp in imputer_opt:
            train_x_imputed, test_x_imputed = get_imputed_data(imp, train_x, test_x)
            score = runClassifier(clf, train_x_imputed, train_y, test_x_imputed, test_y)
            if score > best_score:
                best_score = score
                best_model = name
                best_imputer = imp
        print("******************************************")

print("Best Performance: {0} imputed by {1} with accuracy {2}".format(best_model, best_imputer, best_score))
    
```
